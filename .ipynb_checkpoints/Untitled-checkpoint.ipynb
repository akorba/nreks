```python
import numpy as np
from scipy.stats import norm, expon
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Define the Gaussian distribution function
def gaussian(x, mu, sigma):
    return norm.pdf(x, mu, sigma)

# Define the Exponential distribution function
def exponential(x, scale):
    return expon.pdf(x, scale=scale)

# Define the mixture of distributions alpha*P + (1-alpha)*Q
def mixture_distribution(x, alpha, dist_P, dist_Q):
    return alpha * dist_P(x) + (1 - alpha) * dist_Q(x)

# Define the integrand for the KL divergence
def integrand(x, dist_P, alpha, dist_Q):
    p_x = dist_P(x)
    q_mix_x = mixture_distribution(x, alpha, dist_P, dist_Q)
    if p_x == 0 or q_mix_x == 0:
        return 0.0
    return p_x * np.log(p_x / q_mix_x)

# Function to compute the KL divergence for a given alpha
def compute_kl_divergence(alpha, dist_P, dist_Q):
    kl_divergence, _ = quad(integrand, -np.inf, np.inf, args=(dist_P, alpha, dist_Q))
    return kl_divergence

# Parameters for different distribution pairs
distribution_pairs = [
    {"name": "Gauss (0,2) vs Gauss (1,1)", "P": lambda x: gaussian(x, 0, 2), "Q": lambda x: gaussian(x, 1, 1)},
    {"name": "Gauss (0,1) vs Gauss (1,2)", "P": lambda x: gaussian(x, 0, 1), "Q": lambda x: gaussian(x, 1, 2)},
    {"name": "Gauss (0,1) vs Gauss (-1,1)", "P": lambda x: gaussian(x, 0, 1), "Q": lambda x: gaussian(x, -1, 1)},
    {"name": "Gauss (0,1) vs Expon (1)", "P": lambda x: gaussian(x, 0, 1), "Q": lambda x: exponential(x, 1)},
    {"name": "Gauss (1,1) vs Gauss (1,3)", "P": lambda x: gaussian(x, 1, 1), "Q": lambda x: gaussian(x, 1, 3)},
    {"name": "Expon (1) vs Expon (2)", "P": lambda x: exponential(x, 1), "Q": lambda x: exponential(x, 2)},
]

# Generate values for alpha between 10^-2 and 0.9
alphas = np.linspace(0.01, 0.9, 100)

# Plot results for each pair of distributions
plt.figure(figsize=(12, 8))
for pair in distribution_pairs:
    kl_divergences = [compute_kl_divergence(alpha, pair["P"], pair["Q"]) for alpha in alphas]
    plt.plot(alphas, kl_divergences, label=pair["name"], marker='o')

plt.xlabel('')
plt.ylabel('KL Divergence')
plt.title('KL Divergence between Different Distribution Pairs and Mixture P + (1-)Q')
plt.legend()
plt.grid(True)
plt.show()
```
