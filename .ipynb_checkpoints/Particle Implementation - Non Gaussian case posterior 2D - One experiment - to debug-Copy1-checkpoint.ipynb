### Test case "Gaussian"

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimension
d = 2

# Observed data
y = np.zeros(1)

# Grid boundaries
xmin, xmax = -2, 2
ymin, ymax = -2, 2

# Noise levels
sigNoise = 1
sigPrior = 1

# Define a generic model function G
def G(u):
    return np.array([(u[0] - 1) + (u[1] + 1)**3])  # Example non-linear model

# Compute Jacobian of G numerically using finite differences
def compute_jacobian(G, u, epsilon=1e-5):
    """ Computes the Jacobian matrix of G at u using finite differences. """
    d = len(u)
    J = np.zeros((1, d))  # G outputs 1D, input is d-dimensional
    
    for i in range(d):
        u_plus = u.copy()
        u_minus = u.copy()
        u_plus[i] += epsilon
        u_minus[i] -= epsilon
        
        J[:, i] = (G(u_plus) - G(u_minus)) / (2 * epsilon)  # Central difference
    
    return J

# Define likelihood term
def Phi(u):
    return 0.5 / (sigNoise**2) * (G(u) - y)**2

# Define posterior potential function
def V(u):
    return 0.5 / (sigPrior**2) * np.linalg.norm(u, axis=0)**2 + Phi(u)

# Compute posterior covariance matrix and its inverse
def compute_posterior_covariance():
    u_ref = np.zeros(d)  # Reference point (e.g., prior mean)
    g = compute_jacobian(G, u_ref)  # Compute Jacobian at reference point

    # Compute posterior inverse covariance matrix using Bayesian update rule
    K_inv = np.eye(d, d) * (1 / sigPrior**2) + np.dot(g.T, np.dot(np.eye(1, 1) * (1 / sigNoise**2), g))
    K = np.linalg.inv(K_inv)  # Compute posterior covariance

    return K, K_inv

K, K_inv = compute_posterior_covariance()

print("Posterior Covariance Matrix (K):")
print(K)
print("\nInverse Posterior Covariance Matrix (K_inv):")
print(K_inv)

# Generate 2D grid for visualization
nb_grid = 200
u0s = np.linspace(xmin, xmax, nb_grid)
u1s = np.linspace(ymin, ymax, nb_grid)
U0, U1 = np.meshgrid(u0s, u1s)
U = np.stack((U0, U1), axis=0)  # Shape (2, nb_grid, nb_grid)

# Evaluate the (unnormalized) density function on the grid
def eval_potential(U_grid):
    """ Evaluates I(U) element-wise over a (nb_grid, nb_grid, 2) grid. """
    U_flat = U_grid.reshape(-1, 2)  # Flatten grid for batch processing
    I_vals = np.apply_along_axis(V, 1, U_flat)  # Evaluate I at each point
    return np.exp(-I_vals).reshape(nb_grid, nb_grid)  # Reshape back

unnorm_dens = eval_potential(U.transpose(1, 2, 0))  # Reshape for element-wise processing

# Normalize density
Z = np.trapz(unnorm_dens, u0s, axis=1)
Z = np.trapz(Z, u1s)
dens = unnorm_dens / Z  # normalize density

# Compute marginal distributions
marg_over_x = np.trapz(dens, u0s, axis=1)
marg_over_y = np.trapz(dens, u1s, axis=0)

# Plot posterior distribution 
plt.figure()
DS = plt.contour(U0, U1, dens)  # Plot normalized density
plt.clabel(DS, inline=1, fontsize=10)
labels = ['post'] * len(DS.collections)
for i, label in enumerate(labels):
    DS.collections[i].set_label(label)

# Compare with Gaussian approximation
m = np.zeros(2)
cov_inv = np.linalg.inv(K)
cov_det = np.linalg.det(K)
coe = 1.0 / ((2 * np.pi)**2 * cov_det)**0.5
T = coe * np.exp(-0.5 * (cov_inv[0, 0] * (U0 - m[0])**2 + (cov_inv[0, 1] + cov_inv[1, 0]) * (U0 - m[0]) * (U1 - m[1]) + cov_inv[1, 1] * (U1 - m[1])**2))
CS = plt.contour(U0, U1, T, linestyles="--")  # Gaussian posterior
plt.clabel(CS, inline=1, fontsize=10)
labels = ['reconstruction'] * len(CS.collections)
for i, label in enumerate(labels):
    CS.collections[i].set_label(label)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axis('equal')
plt.title("Posterior Distribution")
plt.show()

def grad_V(u):
    """ Computes the gradient of the posterior potential V(u) for a generic G(u). """
    # Compute the gradient of the prior term
    grad_prior = (1 / sigPrior**2) * u

    # Compute G(u) and its Jacobian numerically
    G_u = G(u)  # Evaluate function G(u)
    grad_G = compute_jacobian(G, u)  # Numerically compute G(u)

    # Compute the gradient of Phi
    grad_Phi = (1 / sigNoise**2) * (G_u - y) @ grad_G  # Matrix multiplication for multi-output G

    # Compute the final gradient of V(u)
    grad_V = grad_prior + grad_Phi

    return grad_V
```
